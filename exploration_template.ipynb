## Univariate Exploration

> I start with duration of trip.

```python
#for avoiding the repeatition
def create_label(title, xlabel, ylabel):
    plt.title(title)
    plt.xlabel(xlabel)
    plt.ylabel(ylabel)
```

```python
binsize = 500
bins = np.arange(0, df.duration_sec.max()+binsize, binsize)

#for sizing the plot
plt.figure(figsize=[14, 7])

#plotting the histogram
plt.hist(data = df, x = 'duration_sec', bins = bins, color='maroon')
create_label('Trip Duration', 'Duration_sec', 'Number of trips')
plt.axis([-500, 10000, 0, 90000]);
```

```python
#As you can see from above plot, tail is too long. Let us use log scale then
log_binsize = 0.05
bins_log = 10 ** np.arange(2.4, np.log10(df.duration_sec.max())+log_binsize, log_binsize)

#for sizing the plot
plt.figure(figsize=[14, 7])

#plotting the histogram
plt.hist(data = df, x = 'duration_sec', bins = bins_log, color='maroon')
create_label('Trip Duration', 'Duration_sec', 'Number of trips')
plt.xscale('log')
plt.xticks([500, 1e3, 2e3, 5e3, 1e4], [500, '1k', '2k', '5k', '10k'])
plt.axis([0, 10000, 0, 15000]);
```

Apparently, trip duration is centered with values which are less than 2000 seconds. The longest trip is around 600-700 seconds. Number of trips starts from around 8000 to 12000 and after that it falls down with 2000 at 2000 seconds.

## Bivariate Exploration

> I want to see the correlations of duration_sec with start_station_name, end_station_name, user_type and rental_access_method.

```python
sorted(df.start_station_id.unique())
```

```python
array_s = []

start_station_ids = sorted(df.start_station_id.unique())
for c in start_station_ids:
    array_s.append(df[df.start_station_id == c].duration_sec.sum())
total = pd.Series(array_s)

plt.figure(figsize=[14,7])
sb.lineplot(x = df.start_station_id, y = total, color = 'Maroon')
plt.xticks(range(0, 540, 20))
create_label('Trip duration and Start station', 'Start station', 'Total duration');
```

```python
array_e = []

end_station_ids = sorted(df.end_station_id.unique())
for c in end_station_ids:
    array_e.append(df[df.end_station_id == c].duration_sec.sum())
total_e = pd.Series(array_e)

plt.figure(figsize=[14,7])
sb.lineplot(x = df.end_station_id, y = total_e, color = 'Maroon')
plt.xticks(range(0, 540, 20))
create_label('Trip duration and End station', 'End station', 'Total duration');
```

As you can see from above two plots, start and end stations of some are higher. In consequence, we can view the longer trips related to start station and end station.

## Multivariate Exploration

> Let us plot the duration_sec, user_type and rental_access_method.

```python
plt.figure(figsize=[14,7])
sb.catplot(data=df, x='user_type', y='duration_sec', hue = 'rental_access_method', kind='violin', height=5, aspect=2)
create_label('Trip duration per User type and Rental access method', 'User type', 'Duration seconds')
plt.ylim(-10, 5000);  
```

```python
plt.figure(figsize=[14,7])
usertype_M = [['Customer', 'o'],['Subscriber', 's']]

for u, m in usertype_M:
    df_Utype = df[df['user_type'] == u]
    plt.scatter(df_Utype['bike_id'], df_Utype['duration_sec'], marker = m, alpha= 1)
    
plt.legend(['Customer','Subscriber'])
plt.axis([10, 3900, -200, 12000 ])
create_label('Trip Duration, user type, and bike_id', 'Bike id', 'Duration_sec');
```

As you can see plot is not clean. So I need to make separate plots

```python
user_type = sb.FacetGrid(data = df, col = 'user_type', col_wrap = 2, size = 5, xlim = [10, 3900], ylim = [-200, 12000])
user_type.map(plt.scatter, 'bike_id', 'duration_sec', alpha= 1)

user_type.set_xlabels('Bike id')
user_type.set_ylabels('Duration_sec');
```
